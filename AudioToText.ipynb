```python
file = "Robert"
audio = AudioSegment.from_wav(file + ".wav")
```

```python
print(audio.duration_seconds)
time = 0
files = []
count = 0
interval = 8000
while time < audio.duration_seconds*1000:
    (audio[max(0, time-200):time+interval]+5).export("file" + str(count) + ".wav", format="wav")
    files.append('file' + str(count) + ".wav")
    time +=interval
    count+=1
```

```python
print(files)
```

```python
def toText(file, r):
        # recognize speech using Google Speech Recognition
    try:
        # for testing purposes, we're just using the default API key
        # to use another API key, use `r.recognize_google(audio, key="GOOGLE_SPEECH_RECOGNITION_API_KEY")`
        # instead of `r.recognize_google(audio)`
        return r.recognize_google(file)
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
    except sr.RequestError as e:
        print("Could not request results from Google Speech Recognition service; {0}".format(e))
```

```python
text = []
count = 0
r = sr.Recognizer()
for f in files:
    r.energy_threshold = 10
    AUDIO_FILE = path.join(path.dirname(path.realpath("Audio")), f)
    with sr.AudioFile(AUDIO_FILE) as source:
        audio = r.record(source)  # read the entire audio file
    google_attempt = toText(audio, r)
    if google_attempt:
        text.append(toText(audio, r) + "*")
    else:
        text.append(google_attempt)
    print(str(count) + " " + str(text[count]))
    count += 1
```

```python
nones = [i for i in range(len(text)) if text[i] == None]
failed_attempts = collections.deque()
[failed_attempts.append(i) for i in range(len(text)) if text[i] == None]
print(nones)
print(len(nones))
print(failed_attempts)
```

```python
# second_pass = []
# count1 = 0
# for index in nones:
#     if (index == 0):
#         continue;
#     f = files[index]
#     print(f)
#     (AudioSegment.from_wav(f) + 10).export("temp.wav", format="wav")
#     AUDIO_FILE = path.join(path.dirname(path.realpath("Audio")), "temp.wav")
#     with sr.AudioFile(AUDIO_FILE) as source:
#         audio = r.record(source)  # read the entire audio file
#     secondTry = toText(audio, r)
#     if secondTry is not None:
#         text[index] = secondTry
#     second_pass.append(secondTry)
#     print(str(index) + " " + str(second_pass[count1]))
#     count1 += 1
    
audio_boost = 10
while len(failed_attempts) > 0.01*len(text) or audio_boost < 60:
    print("audio boost is " + str(audio_boost))
    for i in range(len(failed_attempts)):
        index = failed_attempts.popleft()
        f = files[index]
        print(f)
        (AudioSegment.from_wav(f) + audio_boost).export("temp.wav", format="wav")
        AUDIO_FILE = path.join(path.dirname(path.realpath("Audio")), "temp.wav")
        with sr.AudioFile(AUDIO_FILE) as source:
            audio = r.record(source)  # read the entire audio file
        secondTry = toText(audio, r)
        if secondTry is not None:
            text[index] = secondTry
        else:
            failed_attempts.append(index)
        print(str(index) + " " + str(secondTry))
    audio_boost += 10
    
    
```
